```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import numpy as np 
```

```python
df1 = pd.read_csv('C:/Users/jalen/anaconda/Fitbit/Start.csv')
df2 = pd.read_csv('C:/Users/jalen/anaconda/Fitbit/Recent.csv')
```

```python
df1.info()
```

```python
df1 = df1.drop(columns=['Floors'])
```

```python
df1.info()
```

```python
combined_df = pd.concat([df1, df2])

combined_df.reset_index(drop=True, inplace=True)

print(combined_df)
```

```python
combined_df = combined_df.drop_duplicates()
```

```python
print(combined_df)
```

```python
newdf3 = combined_df.dropna()
newdf3.reset_index(drop=True, inplace=True)
```

```python
print(newdf3)
```

```python
newdf3.info()
newdf3.head()
```

```python
columns_to_convert = ['Steps', 'Calories In', 'Calories Burned', 'Minutes Sedentary', 'Activity Calories']
newdf3_copy = newdf3.copy()
newdf3_copy[columns_to_convert] = newdf3_copy[columns_to_convert].apply(lambda x: x.str.replace(",", "").astype("float64"))
newdf3_cleaned = newdf3_copy.dropna()
newdf3_cleaned.reset_index(drop=True, inplace=True)
```

```python
data = newdf3_cleaned.copy()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.info()
data.head()
```

```python
plt.figure(figsize=(10, 8))

correlation_matrix = data.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap', fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data, vars=['Calories In', 'Calories Burned', 'Steps', 'Distance', 'Minutes Sedentary'])
plt.suptitle('Pair Plot for Multivariate Analysis', fontweight='bold',y=1.02)
plt.tight_layout()
plt.show()
```

```python
start_date = pd.to_datetime('2023-07-10')
end_date = pd.to_datetime('2023-08-30')
filtered_data = data[(data['Date'] >= start_date) & (data['Date'] <= end_date)]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(filtered_data['Date'], filtered_data['Calories In'], label='Calories In')
plt.plot(filtered_data['Date'], filtered_data['Calories Burned'], label='Calories Burned')
plt.xlabel('Date', fontweight='bold')
plt.ylabel('Calories', fontweight='bold')
plt.title('Calories In vs Calories Burned', fontweight='bold')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(filtered_data['Date'], filtered_data['Calories In'], label='Calories In')
plt.bar(filtered_data['Date'], filtered_data['Calories Burned'], label='Calories Burned', alpha=0.4)
plt.xlabel('Date', fontweight='bold')
plt.ylabel('Calories', fontweight='bold')
plt.title('Calories In vs Calories Burned', fontweight='bold')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
numeric_columns = ['Calories In', 'Calories Burned'] 
weekly_data = data.groupby(data['Date'].dt.to_period('W'))[numeric_columns].sum(numeric_only=True)

weekly_data.index = weekly_data.index.astype(str)

plt.figure(figsize=(10, 6))
p1 = plt.bar(weekly_data.index, weekly_data['Calories In'], label='Calories In', alpha=0.7)
p2 = plt.bar(weekly_data.index, weekly_data['Calories Burned'], label='Calories Burned', alpha=0.7, width=0.4)

plt.xlabel('Week', fontweight='bold')
plt.ylabel('Total Calories', fontweight='bold')
plt.title('Total Calories In vs Total Calories Burned by Week', fontweight='bold')

plt.legend(handles=[p1, p2], labels=['Calories In', 'Calories Burned'])

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Steps'], data['Calories Burned'])
plt.xlabel('Steps', fontweight='bold')
plt.ylabel('Calories Burned', fontweight='bold')
plt.title('Scatter Plot of Steps vs Calories Burned', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Steps'], bins=15, edgecolor='black')
plt.xlabel('Steps', fontweight='bold')
plt.ylabel('Frequency', fontweight='bold')
plt.title('Histogram of Steps Distribution', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
activity_columns = ['Minutes Sedentary', 'Minutes Lightly Active', 'Minutes Fairly Active', 'Minutes Very Active']
activity_labels = ['Sedentary', 'Lightly Active', 'Fairly Active', 'Very Active']

activity_distribution = data[activity_columns].sum()

plt.figure(figsize=(8, 8))

explode = (0.05, 0.05, 0.2, 0.1) 

wedgeprops = {'linewidth': 1, 'edgecolor': 'white', 'alpha': 0.7}

plt.pie(activity_distribution, labels=activity_labels, autopct='%1.1f%%', startangle=140, explode=explode, shadow=False, wedgeprops=wedgeprops, textprops={'fontsize': 7})
plt.title('Activity Minutes Distribution', fontsize=16, pad=26, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
df3 = pd.read_csv('C:/Users/jalen/anaconda/Fitbit/Sleep.csv')
```

```python
df3.info()
```

```python
df3.head()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')

num_only = df3.drop("Start Time", axis = 1)
```

```python
num_only.info()
```

```python
num_only1 = num_only.drop("End Time", axis = 1)
```

```python
num_only1.info()
```

```python
imputer.fit(num_only1)
```

```python
imputer.statistics_
```

```python
num_only1.median().values
```

```python
X = imputer.transform(num_only1)
```

```python
sleep = pd.DataFrame(X, columns=num_only1.columns, index=num_only1.index)
```

```python
sleep.info()
```

```python
df3['Start Time'] = pd.to_datetime(df3['Start Time'])
df3['End Time'] = pd.to_datetime(df3['End Time'])

imputed_sleep_data = pd.concat([df3[['Start Time', 'End Time']], sleep], axis=1)

print(imputed_sleep_data.head())
```

```python
average_sleep_duration = imputed_sleep_data['Minutes Asleep'].mean()/60
print(f"Average Sleep Duration: {average_sleep_duration:.2f} Hours")
```

```python
imputed_sleep_data['Sleep Efficiency'] = (imputed_sleep_data['Minutes Asleep'] / imputed_sleep_data['Time in Bed']) * 100
average_sleep_efficiency = imputed_sleep_data['Sleep Efficiency'].mean()
print(f"Average Sleep Efficiency: {average_sleep_efficiency:.2f}%")
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = imputed_sleep_data.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation between Sleep Variables', fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=imputed_sleep_data, x= 'Number of Awakenings', y=imputed_sleep_data['Minutes Asleep']/60, hue='Sleep Efficiency', palette='coolwarm', size='Sleep Efficiency', alpha=0.75, sizes=(80, 500))
plt.xlabel('Number of Awakenings', fontweight='bold')
plt.ylabel('Hours Asleep', fontweight='bold')
plt.title('Awakenings vs. Sleep Duration', fontweight = 'bold')
plt.legend(title='Sleep Efficiency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(imputed_sleep_data['Start Time'], imputed_sleep_data['Minutes Asleep']/60, marker='o')
plt.xlabel('Date', fontweight='bold')
plt.ylabel('Hours Asleep', fontweight='bold')
plt.title('Sleep Duration Over Time',fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
stage_distribution = imputed_sleep_data[['Minutes REM Sleep', 'Minutes Light Sleep', 'Minutes Deep Sleep']].sum()
sleep_stages = stage_distribution.index

plt.figure(figsize=(8, 8))
colors = ['skyblue', 'lightgreen', 'lightcoral']
plt.pie(stage_distribution, labels=sleep_stages, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Sleep Stages Distribution', fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=imputed_sleep_data, x=imputed_sleep_data['Minutes Asleep']/60, bins=15, kde=True)
plt.xlabel('Hours Asleep', fontweight='bold')
plt.ylabel('Frequency', fontweight='bold')
plt.title('Sleep Duration Distribution', fontweight='bold')
plt.grid(True)
plt.tight_layout()
plt.show()
```
